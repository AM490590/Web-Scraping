```python
import requests
from bs4 import BeautifulSoup
import csv

with open('shagoof2.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Price'])

    for page in range(4):
        response = requests.get(f"https://shagoof.net/collections/all?page={page}")
        response.status_code  # Check the response status

        soup = BeautifulSoup(response.content, 'html.parser')
        # Assuming the product elements are in divs with a specific class
        products = soup.find_all('div', class_='t4s-product-wrapper') # Find all product divs

        for product in products:
            name = product.find('h3').text.strip()
            price = product.find('div', class_="t4s-product-price").text.strip()
            writer.writerow([name, price])
```
